5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

n_samples = 1000 
X = np.random.randn(n_samples, 3)


feature_4 = 0.5 * X[:, 0] + 0.3 * X[:, 1] + np.random.randn(n_samples) * 0.1

feature_5 = X[:, 2] + np.random.randn(n_samples) * 0.1

data = pd.DataFrame(np.column_stack([X, feature_4, feature_5]),
                    columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5'])

corr_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True, linewidths=0.5)

plt.title('Correlation Matrix of Features')

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(20)

# Generate the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the data to create a matrix of Sales by Month and Day
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Set up the figure and axis for the heatmap
plt.figure(figsize=(12, 8))

# Create the heatmap
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5, cbar_kws={'label': 'Average Sales'})

# Add title and labels
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(25)

data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),  # Random population between 100 and 1000 million
    'GDP': np.random.randint(500, 2000, 5)  # Random GDP between 500 and 2000 billion USD
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

plt.scatter(
    df['GDP'], 
    df['Population'], 
    s=df['Population'] * 10,  # Bubble size proportional to population (scaled)
    c=df['Population'],  # Color based on population
    cmap='viridis',  # Colormap
    alpha=0.6,  # Transparency
    edgecolors='w',  # White edge color for bubbles
    linewidth=1.5  # Edge line width
)

for i, country in enumerate(df['Country']):
    plt.text(
        df['GDP'][i], 
        df['Population'][i], 
        country, 
        fontsize=12, 
        ha='center', 
        va='center', 
        color='black'
    )

plt.title('Bubble Chart: Population vs GDP by Country', fontsize=14)
plt.xlabel('GDP (Billion USD)', fontsize=12)
plt.ylabel('Population (Millions)', fontsize=12)

# Add color bar for the population
plt.colorbar(label='Population')

# Show plot
plt.grid(True)
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file

x = np.linspace(0, 10, 100)
y = np.sin(x)

output_file("sine_wave.html")

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

show(p)
```

5. Create a Bokeh heatmap using the provided dataset.               
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.palettes import Viridis256
from bokeh.transform import linear_cmap

data_heatmap = np.random.rand(10, 10)  
x = np.linspace(0, 1, 10) 
y = np.linspace(0, 1, 10) 
xx, yy = np.meshgrid(x, y) 

xx = xx.flatten()
yy = yy.flatten()
values = data_heatmap.flatten()

data = {
    'x': xx,
    'y': yy,
    'value': values
}

output_file("heatmap.html")

p = figure(title="Bokeh Heatmap", x_axis_label='X', y_axis_label='Y', 
           width=800, height=800, toolbar_location=None)

mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(values), high=max(values))

p.rect(x='x', y='y', width=0.1, height=0.1, source=data, color=mapper, line_color=None)

p.xgrid.grid_line_color = None  
p.ygrid.grid_line_color = None  
p.xaxis.visible = False  
p.yaxis.visible = False  

show(p)
```
